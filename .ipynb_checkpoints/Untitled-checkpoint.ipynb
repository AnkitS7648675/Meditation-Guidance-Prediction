### Data Preprocessing

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=3000)
X_text = tfidf.fit_transform(df['journal_text'])
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = ['sleep_hours', 'energy_level', 'stress_level', 'duration_min']

X_num = df[num_cols].fillna(0)

scaler = StandardScaler()
X_num_scaled = scaler.fit_transform(X_num)
```

```python
from scipy.sparse import hstack
X = hstack([X_text, X_num_scaled])
```

### Emotional State Prediction

```python
y_state = df['emotional_state']
```

```python
from sklearn.linear_model import LogisticRegression

model_state = LogisticRegression(max_iter=1000)
model_state.fit(X, y_state)
```

```python
X = tfidf.fit_transform(df['journal_text'])
y_state = df['emotional_state']

model_state.fit(X, y_state)
```

### Decision Engine

```python
def decide_action(state, intensity, stress, energy, time):
    
    if intensity >= 4 and stress > 7:
        return "breathing", "now"
    
    if state in ["sad", "low"] and energy < 4:
        return "rest", "later_today"
    
    if state == "anxious":
        return "grounding", "within_15_min"
    
    if energy > 7:
        return "deep_work", "now"
    
    return "journaling", "tonight"
```
